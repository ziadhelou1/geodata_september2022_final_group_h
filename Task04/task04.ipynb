```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import datetime
dateparse = lambda dates: [datetime.strptime(str(d), '%Y%m%d') for d in dates]
```

```python
columns = ['Date_reported', 'Country', 'New_cases', 'New_deaths']
#columns that we need to show in our data
```

```python
#Extracting only the columns needed from the excel
df_data = pd.read_csv(r'data/WHO-COVID-19-global-data.csv', usecols=columns,
                      parse_dates = ['Date_reported'])
df_data
```

```python
#grouping the data into weekly data to be lighter on QGIS (from 234156 rows to 33654) 
df_data1 = (df_data.reset_index().set_index("Date_reported")
            .groupby(["Country",pd.Grouper(freq='W-MON')])['New_cases','New_deaths']
            .sum().astype(int).reset_index())
df_data1
```

```python
#sorting based on the dates
df_data2 = df_data1.sort_values('Date_reported')
columns_titles = ["Date_reported","Country", "New_cases", "New_deaths"]
df_covid19 =df_data2.reindex(columns=columns_titles)
df_covid19
```

```python
#downloading the data csv
df_covid19.to_csv(r"data/covid19_data.csv")
```
